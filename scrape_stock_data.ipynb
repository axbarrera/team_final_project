```python
import pandas as pd
import string
```

```python
letter_list = list(string.ascii_lowercase)
df = pd.DataFrame(columns = ['Abbr','Company'])
for letter in letter_list:
    url = 'http://eoddata.com/stocklist/NASDAQ/%s.htm' %letter
    print('Scraping:',url)
    html = pd.read_html(url)
    lettered_df = html[4][[0,1]]
    lettered_df = lettered_df[1:]
    lettered_df = lettered_df.rename(columns={0:'Abbr',1:'Company'})
    df = df.append(lettered_df).reset_index(drop=True)
```

```python
with open('name_table.sql','w') as f:
    f.write("drop table if exists NAMES;\n")
    f.write("CREATE table NAMES (Abbr varchar(10),Company varchar(100));\n")
    f.write('insert into NAMES (Abbr,Company)\n')
    f.write('values')
    for i in range(0,(len(df) - 2)):
        output = '(' + '"' + df.iat[i,0] + '"' + ',' + '"' + df.iat[i,1] + '"' + '),\n'
        f.write(output)
    f.write('(' + '"' + df.iat[len(df) - 1,0] + '"' + ',' + '"' + df.iat[len(df) - 1,1] + '"' + ");")
```
